### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
%matplotlib inline
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd


sales = pd.read_csv('./data/company_sales_data.csv', sep = ',')

        
plt.plot(sales["month_number"], sales["total_profit"])
plt.title("Total Profit of All Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
````

### Exercise 2. 

Get Total Units Sold of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Units Sold Number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
%matplotlib inline
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd


sales = pd.read_csv('./data/company_sales_data.csv', sep = ',')
       
    
plt.plot(sales["month_number"], sales["total_units"], label='Total Units', color='r', linestyle=':', linewidth='3', marker='o')
plt.title("Total Units Sold of All Months")
plt.xlabel("Month Number")
plt.ylabel("Units Sold Number")
plt.legend(loc='lower right')
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
%matplotlib inline
import matplotlib.pyplot as plt

#plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

sales = pd.read_csv('./data/company_sales_data.csv', sep = ',')
       
    
plt.plot(sales["month_number"], sales["facecream"], label='Face Cream', color='r', linestyle='solid', linewidth='3', marker='o')
plt.plot(sales["month_number"], sales["facewash"], label='Face Wash', color='b', linestyle=':', linewidth='3', marker='o')
plt.plot(sales["month_number"], sales["toothpaste"], label='Toothpaste', color='g', linestyle='dashed', linewidth='3', marker='o')
plt.plot(sales["month_number"], sales["bathingsoap"], label='Bathing Soap', color='y', linestyle='dashdot', linewidth='3', marker='o')
plt.plot(sales["month_number"], sales["shampoo"], label='Shampoo', color='#9932CC', linestyle=':', linewidth='3', marker='o')
plt.plot(sales["month_number"], sales["moisturizer"], label='Moisturizer', color='#FFA500', linestyle='dashed', linewidth='3', marker='o')


plt.title("All Product Sales")
plt.xlabel("Month Number")
plt.ylabel("Units Sold Number")
plt.legend(loc='upper left')
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
%matplotlib inline
import matplotlib.pyplot as plt
#plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

sales = pd.read_csv('./data/company_sales_data.csv', sep = ',')

plt.bar(x = sales["month_number"], height = sales["facecream"], width = .25)
plt.bar(x = sales["month_number"]+.25, height = sales["facewash"], width = .25)


plt.title("Face Cream and Face Wash Sold")
plt.xlabel("Month Number")
plt.ylabel("Units Sold Number")
````
